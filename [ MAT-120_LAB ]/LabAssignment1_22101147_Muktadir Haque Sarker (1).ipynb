```python
# TASK - 01

import matplotlib.pyplot as plt
import numpy as np
from sympy import symbols, N, log


interval = [-5, 6]

x = symbols("x")
y = 9*x**3 -7*x**2 +3*x+10
print(y)
df1 = diff(y, x)
eq1 = Eq(df1, 0)
solu = solve(eq1)
domran = []
for i in solu:
    domran.append([i, y.subs(x, i)])
domran2 = []
for i in interval:
    domran2.append([i, y.subs(x, i)])
final = domran + domran2

maxima = str("No Maxima")
minima = str("No Minima")
for i in range(len(final)):
    try:
        if final[i][-1] > 0 :
            maxima = tuple(final[i])
        elif final[i][-1] < 0:
            minima = tuple(final[i])
    except:
        continue
print("Maxima :", maxima)
print("Minima :", minima)

dif1 = lambdify(x, y)
points = np.arange(-5, 6)
a = dif1(points)
plt.title("shape of function")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.plot(points, a)
plt.show()
```
