```python
# TASK - 1

class Customer:
    def __init__(self, name):
        self.name = name
        self.gname = str()
        self.items = tuple()
    def purchase(self, *items):
        self.items = items
        print(f'{self.name}, you purchased {len(self.items)} item(s):')
        for i in self.items:
            print(i)
    def greet(self, gname = ""):
        self.gname = gname+"!"
        if gname == "":
            print("Hello!")
        else:
            print("Hello "+str(self.gname))
        

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
# Task - 2

class Panda:
    def __init__(self, name, gender, age):
        self.name = name
        self.gender = gender
        self.age = age
    def sleep(self, time = None):
        if time != None:
            if 3 <= time <= 5:
                plan = "Mixed Veggies"
            elif 6 <= time <= 8:
                plan = "Eggplant & Tofu"
            elif 9 <= time <= 11:
                plan = "Broccoli Chicken"
        else:
            plan = "bamboo leaves"
            return f"{self.name}'s duration is unknown thus should have only {plan}"
        return f"{self.name} sleeps {time} hours daily and should have {plan}"


panda1 = Panda("Kunfu","Male", 5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)
print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
# Task - 3

class Cat:
    def __init__(self, col1 = "White", stat = None):
        if stat == None:
            self.col1 = col1
            self.output = f'{self.col1} cat is sitting'
        else:
            self.col1 = col1
            self.stat = stat
            self.output = f'{self.col1} cat is {self.stat}'
    def changeColor(self, col):
        if col == "Blue":
            self.output = f'{col} cat is sitting'
        else:
            self.output = f'{col} cat is jumping'
    def printCat(self):
        print(self.output)
    
c1 = Cat()
c2 = Cat("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
```

```python
# Task - 5

class Student:
    def __init__(self, name, id, dep="CSE"):
        self.name = name
        self.id = id
        self.dep = dep
        self.time = 0
        self.sug = "Suggestion:"
    def dailyEffort(self, time):
        self.time = time 
        if time <= 2:
            self.sug += " Should give more effort!"
        elif time <= 4:
            self.sug += " Keep up the good work!'"
        else:
            self.sug += " Excellent! Now motivate others."    
    def printDetails(self):
        print("Name:",self.name)
        print("ID:", self.id)
        print("Department:", self.dep)
        print("Daily Effort:",self.time,"hour(s)")
        print(self.sug)
        
harry = Student('Harry Potter', 123)
harry.dailyEffort(3)
harry.printDetails()
print('========================')
john = Student("John Wick", 456, "BBA")
john.dailyEffort(2)
john.printDetails()
print('========================')
naruto = Student("Naruto Uzumaki", 777, "Ninja")
naruto.dailyEffort(6)
naruto.printDetails()
```

```python
# Task - 7

class Match:
    def __init__(self, match):
        print("5..4..3..2..1.. Play !!!")
        self.team = match.split("-")
        self.bat, self.ball = self.team[0], self.team[1]
        self.run = 0
        self.over = 0
        self.wicket = 0
    def add_run(self, run):
        self.run += run
    def add_over(self, over):
        if over >= 5:
            print("Warning! Cannot add 5 over/s (5 over match)")
            pass
        else:
            self.over += over
    def add_wicket(self, wicket):
        self.wicket = wicket
    def print_scoreboard(self):
        return f'Batting Team: {self.bat}\nBowling Team: {self.ball}\nTotal Runs: {self.run} Wickets: {self.wicket} Overs: {self.over}'

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
# Task - 8

class ParcelKoro:
    def __init__(self, name = "No name set", weight = 0):
        self.name = name
        self.product_weight = weight
        self.fee = 0
    def calculateFee(self, place = None):
        if place == None and self.product_weight != 0:
            lc = 50
        elif self.product_weight != 0:
            lc = 100
        else:
            lc = 0
        self.fee = (self.product_weight * 20) + lc
    def printDetails(self):
        print("Customer Name:", self.name)
        print("Product Weight:", self.product_weight)
        print("Total fee:", self.fee)
    

    
print("**********************")
p1 = ParcelKoro()
p1.calculateFee()
p1.printDetails()
print("**********************")
p2 = ParcelKoro('Bob The Builder')
p2.calculateFee()
p2.printDetails()
print("----------------------------")
p2.product_weight = 15
p2.calculateFee()
p2.printDetails()
print("**********************")
p3 = ParcelKoro('Dora The Explorer', 10)
p3.calculateFee('Dhanmondi')
p3.printDetails()
```

```python
# Task - 9

class Batsman:
    def __init__(self, a, b, c=None):
        if c != None:
            self.name = a
            self.rs = b
            self.bf = c
        else:
            self.name = "New Batsman"
            self.rs = a
            self.bf = b
    def printCareerStatistics(self):
        print("Name:", self.name)
        print(f"Runs Scored: {self.rs} , Balls Faced: {self.bf}")
    def battingStrikeRate(self):
        return self.rs / self.bf * 100
    def setName(self, name):
        self.name = name


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
# Task - 12

class TaxiLagbe:
    def __init__(self, info, place):
        self.info = info 
        self.place = place
        self.passen = []
        self.man = []
        self.fare = 0
    def addPassenger(self, *passen):
        temp = []
        temp += list(passen)
        self.passen = [x.split("_") for x in temp]
        for i in range(len(self.passen)):
            if len(self.man) < 4:
                self.man.append(self.passen[i][0])
                self.fare += int(self.passen[i][1])
                for i in self.passen:
                    print(f'Dear {i[0]}! Welcome to TaxiLagbe.')
            else:
                print("Taxi Full! No more passengers can be added.")
    def printDetails(self):
        print(f'Trip info for Taxi number: {self.info}')
        print(f'This taxi can cover only {self.place} area.')
        print(f'Total passengers: {len(self.man)}')
        print("Passenger lists:")
        print(*self.man, sep=", ")
        print(f'Total collected fare: {self.fare} Taka')
        
        
    
        
    



taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
# Task - 13

class Account:
    def __init__(self, name = "Default Account", balance = 0.00):
        self.name = name
        self.balance = balance
    def details(self):
        return f'{self.name}\n{self.balance}'
    def withdraw(self, amu):
        newbal = self.balance - amu
        if newbal <= 3070:
            print("Sorry, Withdraw unsuccessful! The account balance after deducting withdraw amount is equal to or less than minimum.")
        else:
            print(f'Withdraw successful! New balance is: {newbal}')


a1 = Account()
print(a1.details())
print("------------------------")
a1.name = "Oliver"
a1.balance = 10000.0
print(a1.details())
print("------------------------")
a2 = Account("Liam")
print(a2.details())
print("------------------------")
a3 = Account("Noah",400)
print(a3.details())
print("------------------------")
a1.withdraw(6930)
print("------------------------")
a2.withdraw(600)
print("------------------------")
a1.withdraw(6929)
```

```python
# Task - 14

class StudentDatabase:
    def __init__(self, name, id):
        self.name = name
        self.id = id
        self.grades = {}
    def calculateGPA(self, grades, semester):
        sum = 0
        courses = []
        for items in grades:
            item = items.split(":")
            course = item[0].strip()
            grade = float(item[1].strip())
            courses.append(course)
            sum += grade
            cg = round(sum / len(grades),2)
        self.grades[semester] = {tuple(courses) : cg}
    def printDetails(self):
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        for sem in self.grades:
            print(f"Courses taken in {sem}:")
        for courses in self.grades[sem].keys():
            for course in courses:
                print(course)
        print(f"GPA: {list(self.grades[sem].values())[0]}")



s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'],
'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'],
'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```
