```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  
    # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    def __init__(self, a):
        if(type(a) == list):
            self.head = Node(a[0],None, None)
            temp = self.head
            for i in range(1, len(a)):
                newnode = Node(a[i], None, temp)
                temp.next = newnode
                temp = temp.next
            self.tail = temp
            temp.next = self.head
            self.head.prev = temp
    
    # Counts the number of Nodes in the list and return the number
    def countNode(self):
        temp = self.head
        cnt = 0
        while temp != self.tail:
            cnt += 1
            temp = temp.next
        return 1+cnt
    
    # prints the elements in the list
    def forwardprint(self):
        temp = self.head
        while temp != self.tail:
            print(temp.element, end=" ")
            temp = temp.next
        print(temp.element)
            


    # prints the elements in the list backward
    def backwardprint(self):
        temp = self.tail
        while temp != self.head:
            print(temp.element, end=" ")
            temp = temp.prev
        print(temp.element)

    # returns the reference of the at the given index. For invalid index return None.
    def nodeAt(self, idx):
        temp = self.head
        ind = 0
        while ind < self.countNode():
            temp = temp.next
            ind += 1
            if(ind == idx):
                return temp
        return None
    
    # returns the index of the containing the given element. if the element does not exist in the List, return -1.
    def indexOf(self, elem):
        temp = self.head
        ind = 0
        while temp != None:
            temp = temp.next
            ind += 1
            if(temp.element == elem):
                return ind
        return -1

    # inserts containing the given element at the given index Check validity of index. 
    def insert(self, elem, idx):
        if(idx > self.countNode()):
            print("Invalid")
            return
        newnode = Node(elem, None, None)
        if(idx == 0):
            temp = self.head
            newnode.prev = self.tail
            self.tail.next = newnode
            newnode.next = self.head
            self.head.prev = newnode
            self.head = newnode
            return
        if(idx == self.countNode()):
            self.tail.next = newnode
            newnode.prev = self.tail
            self.tail = newnode
        else:
            temp = self.head
            ind = 0
            while ind < idx-1:
                temp = temp.next
                ind += 1
            newnode.prev = temp
            newnode.next = temp.next

            temp.next.prev = newnode
            temp.next = newnode
            return
        
        
    
    # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
    def remove(self, idx):
        if(idx > self.countNode()):
            return None
        temp = self.head
        if(idx == 0):
            rem = self.head.element
            self.head = temp.next
            self.head.prev = self.tail
            self.tail.next= self.head
            del(temp)
            return str(rem)
        elif(idx == self.countNode()-1):
                rem = self.tail.element
                self.tail = self.tail.prev
                self.tail.next = self.head
                self.head.prev = self.tail
                return str(rem)
        else:
            ind = 0
            while ind < idx-1:
                temp = temp.next
                ind += 1
            rem = temp.next.element 
            temp.next.next.prev = temp
            temp.next = temp.next.next
            return str(rem)
            
                
            
```
